```python
import requests
import re
import pandas as pd
import time
import numpy as np
```

```python
def get_scores(id):
    r = requests.get(f"https://returnyoutubedislikeapi.com/votes?videoId={id}")
    vals = r.json()
    likes, dislikes, viewCount = vals["likes"], vals["dislikes"],vals["viewCount"]
    return likes,dislikes, viewCount
```
